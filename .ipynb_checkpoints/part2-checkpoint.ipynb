## Importing Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

## Loading In Data Set

```python
colnames=['age', 'sex', 'chest_pain', 'resting_bp', 'cholesterol', 'fasting_bs', 'resting_ecg', 'max_hr', 'exercise_angina', 'oldpeak', 'slope_of_ST segment', '#_of_major_vessels', 'thal', 'Disease_Presence']

heart = pd.read_csv('data/heart.dat', delimiter=' ', index_col=False, names=colnames)
```

## First-Pass Model

```python
# Split the data into feature columns and target column
target_col = 'Disease_Presence'
y = heart.loc[:, target_col]
X = heart.drop(['Disease_Presence'], axis='columns')
```

```python
# Do train/test split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=11)

# Train a logistic regression estimator
lr = LogisticRegression()
lr.fit(X_train, y_train)

# Calculate test-set accuracy
lr.score(X_test, y_test)
```

```python
# Check class frequencies
y_test.value_counts(normalize=True)
```

```python
# Get the confusion matrix for our admissions model
y_pred = lr.predict(X_test)
metrics.confusion_matrix(y_test, y_pred)
```

```python
#Measure the model's accuracy
accuracy = (39+20)/(39+4+5+20)
precision = (20/24)
recall = (20/25)

print(accuracy)
print(precision)
print(recall)
```

This model seems to perfrom significantly better than the null. If you were to guess that nobody in this sample had a heart disease you would be correct about 63% of the time. On the other hand, using my model would predict whether or not they had a heart disease correctly about 87% of the time.
